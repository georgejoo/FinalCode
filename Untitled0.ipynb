```python

import numpy as np
import matplotlib.pyplot as plt

class SinusoidalObserver:
    def __init__(self, dt, g_b, g_a, g_w, g_theta):
        self.dt = dt
        # Adaptation gains
        self.g_b = g_b
        self.g_a = g_a
        self.g_w = g_w
        self.g_theta = g_theta

        # Initial states
        self.b_hat = 0.0
        self.a_hat = 1.0
        self.w_hat = 0.0
        self.theta_hat = 0.0

    def update(self, y):
        # 1. Calculate estimated signal and error
        y_hat = self.a_hat * np.sin(self.theta_hat) + self.b_hat
        e = y - y_hat

        # 2. Update parameters using Gradient Descent (Euler Integration)
        self.b_hat += self.dt * (self.g_b * e)
        self.a_hat += self.dt * (self.g_a * e * np.sin(self.theta_hat))

        # Frequency and Phase updates (PLL structure)
        dw = self.g_w * e * np.cos(self.theta_hat)
        self.w_hat += self.dt * dw

        d_theta = self.w_hat + self.g_theta * e * np.cos(self.theta_hat)
        self.theta_hat += self.dt * d_theta

        # Wrap theta to [0, 2*pi] for numerical stability
        self.theta_hat = self.theta_hat % (2 * np.pi)

        return y_hat, self.b_hat, self.a_hat, self.w_hat

# --- Simulation ---
dt = 0.0001  # 1ms sampling
t = np.arange(0, 5.0, dt)

# True parameters (with a step change at t=1.0)
def get_true_signal(t):
    b = 2.0 if t < 1.0 else 1.0
    a = 5.0
    f = 5 # 10 Hz
    w = 2 * np.pi * f
    return a * np.sin(w * t) + b, b, a, w

# Initialize Observer
# Gains must be tuned based on noise and required convergence speed
obs = SinusoidalObserver(dt, g_b=20, g_a=20, g_w=2000, g_theta=100)

# Data storage
results = {'y': [], 'y_hat': [], 'b_hat': [], 'a_hat': [], 'w_hat': []}

for ti in t:
    y_true, b_true, a_true, w_true = get_true_signal(ti)
    # Add some noise
    y_meas = y_true + np.random.normal(0, 0.2)

    y_h, b_h, a_h, w_h = obs.update(y_meas)

    results['y'].append(y_meas)
    results['y_hat'].append(y_h)
    results['b_hat'].append(b_h)
    results['a_hat'].append(a_h)
    results['w_hat'].append(w_h)

# --- Plotting ---
plt.figure(figsize=(10, 8))
plt.subplot(3, 1, 1)
plt.plot(t, results['y'], label='Measured', alpha=0.5)
plt.plot(t, results['y_hat'], label='Estimated', color='red')
plt.title('Signal Tracking')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, results['b_hat'], label='Bias Estimate')
plt.axhline(2.0, color='k', linestyle='--', alpha=0.3)
plt.plot(t, results['a_hat'], label='Gain Estimate')
plt.axhline(5.0, color='k', linestyle='--', alpha=0.3)
plt.title('Parameter Convergence')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, np.array(results['w_hat'])/(2*np.pi), label='Freq Estimate (Hz)')
plt.axhline(10.0, color='k', linestyle='--', alpha=0.3)
plt.title('Frequency Tracking')
plt.legend()
plt.tight_layout()
plt.show()
```
